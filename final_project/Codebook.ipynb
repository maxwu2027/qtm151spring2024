```python
import pandas as pd

# Load datasets
drivers = pd.read_csv('final_project/data_raw/drivers.csv')
races = pd.read_csv('final_project/data_raw/races.csv')
results = pd.read_csv('final_project/data_raw/results.csv')
circuits = pd.read_csv('final_project/data_raw/circuits.csv')

# Merge operations
race_results = pd.merge(races, results, on='raceId')
driver_results = pd.merge(race_results, drivers, on='driverId')
final_data = pd.merge(driver_results, circuits, on='circuitId')

final_data

```

```python
# Data cleaning code example
final_data['date'] = pd.to_datetime(final_data['date'])
final_data.dropna(subset=['fp1_time'], inplace=True)
final_data = final_data[final_data['year'] >= 2000]

final_data
```

```python
race_results = pd.merge(races, results, on='raceId', how='inner')

# Next, incorporate driver information to connect drivers with their race results
driver_race_results = pd.merge(race_results, drivers, on='driverId', how='inner')

# Finally, add circuit information to analyze the influence of circuit characteristics
complete_data = pd.merge(driver_race_results, circuits, on='circuitId', how='inner')

# Filtering records for modern Formula 1 races
complete_data = complete_data[complete_data['year'] >= 2000]
```

```python
# Filling missing 'time' and 'milliseconds' values with median
complete_data['milliseconds'].fillna(complete_data['milliseconds'].median(), inplace=True)
```
